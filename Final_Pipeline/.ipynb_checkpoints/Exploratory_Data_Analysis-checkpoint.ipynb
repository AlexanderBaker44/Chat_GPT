```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the CSV file from the data folder and set index to 0
fred_data = pd.read_csv('data/fred_data_with_index.csv', index_col=0)

fred_data = fred_data.rename(columns={'GDPC1': 'Real_GDP',
                                      'CPIAUCSL': 'CPI',
                                      'UNRATE': 'Unemployment_Rate',
                                      'FEDFUNDS': 'Federal_Funds_Rate',
                                      'CIVPART': 'Labor_Force_Participation_Rate',
                                      'INDPRO': 'Industrial_Production',
                                      'TOTALSA': 'Vehicle_Sales',
                                      'DSPIC96': 'Disposable_Income',
                                      'PCE': 'Personal_Consumption_Expenditures',
                                      'DGS10': '10Y_Treasury_Rate'})

fred_data = fred_data.rename(columns={'CPALTT01USM657N': 'Adjusted_CPI'})

# Convert the index to a datetime index
fred_data.index = pd.to_datetime(fred_data.index)

import matplotlib.pyplot as plt

# loop through each column in fred_data
for col in fred_data.columns:
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.plot(fred_data[col].dropna())
    ax.set_title(col)
    plt.show()

```

```python
fred_data['Real_GDP'] = fred_data['Real_GDP'].interpolate(method='linear')
```

Let's take a look at our interpolation result, it seems to follow the trends pretty similarly

- Looking at the smoothing values for each of the line charts, many of the major perturbances were smoothed out.
- The general trends become more apparent, but this exponential smoothing leaves out alot of information. 
- These plots tend to reinforce the trends of the previous line charts, however it may be better to look at the differences over time


```python
import pandas as pd
import matplotlib.pyplot as plt

# set the span parameter for exponential smoothing
span = 20

# loop through each column in fred_data
for col in fred_data.columns:
    # apply exponential smoothing to the column
    smoothed_col = fred_data[col].ewm(span=span).mean()

    # plot the original data and the smoothed data on the same chart
    plt.plot(fred_data[col], label=col)
    plt.plot(smoothed_col, label=f'{col} (Smoothed)')

    # add labels and legend to the chart
    plt.title(f'Exponentially Smoothed {col}')
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.legend()

    # show the chart
    plt.show()
```

Now let's take a look at the percent changes over time
- Real_GDP has small changes until 2020 followed by a major negative and positive change
- CPI has many major changes with a major negative change in 2014 and 2020 positive spikes occur after
- Unemployment has small changes until 2020 with a huge spike in this year
- Federal Funds have a spike in 2016, negative spike in 2020 and a postive spike in 2022
- Labor Force Participation has small changes until 2020, a massive downward spike followed by a positive spike
- Industrial production maintains steady with a huge downturn in 2020 and a large positive spike following
- Vehice Sales maintain steady until a downword spike in 2020 and a positive spike in 2021
- Disposable income maintains steady with small downward spike until huge fluctuations in the 2020-2021 period
- Personal consumption maintains a steady rate until massive fluctuations occur in 2020-2021. down spikes followed by two upspikes
- The 10 Year Treasury Rate Changes are consistently fluctuating with numerous large up and down spikes
- The Adjusted CPI maintains consistent until a massive spike in 2020

```python
import pandas as pd
import matplotlib.pyplot as plt

# loop through each column in fred_data
for col in fred_data.columns:
    # calculate the percentage change between each period
    changes = fred_data[col].pct_change()

    # plot the changes on a line chart
    plt.plot(changes, label=col)

    # add labels and legend to the chart
    plt.title(f'{col} Changes')
    plt.xlabel('Date')
    plt.ylabel(f'{col} % Change')
    plt.legend()

    # show the chart
    plt.show()
```

To look further at the orders:
- Real_GDP demonstrates a drop off for the PACF after th 1st lag consistent drop for ACF
- CPI has a drop off for the PACF after the first lag, consistent drop off for the ACF
- Unemplyment Rate has a dropoff after the 1st lag and consistent drop off for the acf plot
- Federal Funds rate has a drop off after the first lag within the PACF and a consistent dropoff for the ACF
- Labor Force Participation Rate has a drop off after the 1st lag for the PACF and a consistent drop off for the ACF
- For industrial production you can see the loss of significance after the 2nd lag, and a consistent decrease for the ACF plot
- Within Vehicle Sales there is a drop off after the first lag for the PACF plot and a general drop off for the ACF plot, however, you can see the semblance of a slight seasonality pattern
- For Dispoabale Income, you can see the drop off after the second lag for the PACF plot, and a consistent decrease within the ACF plot
- For the Personal Consumption Expenditures, there is a drop off after the first lag for the PACF and a gradual decrease within the ACF plot
- For the 10Y Treasury Rate there is a dropoff after the 1st lag for the PACF plot, and a gradual decrease for the ACF plot
- For the Adjusted CPI You see drop offs after the 1st lag for both the ACF and APCF plots. You can also see some semblance of seasonality in these plots. 

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# loop through each column in fred_data
for col in fred_data.columns:
    fig, axes = plt.subplots(1, 2, figsize=(10,3))
    # plot autocorrelation
    plot_acf(fred_data[col].dropna(), ax=axes[0])
    axes[0].set_title(f'Autocorrelation Plot of {col}')
    # plot partial autocorrelation
    plot_pacf(fred_data[col].dropna(), ax=axes[1])
    axes[1].set_title(f'Partial Autocorrelation Plot of {col}')
    plt.tight_layout()
    plt.show()
```
